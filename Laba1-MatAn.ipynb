```python
import matplotlib.pyplot as plt
import numpy as np
import math
from random import randint
```

```python
def Func(x): #Задаём функцию.
    global F
    F = ''
    if F == 'cosx': 
        return x**8/(8*7*6*5*4*3*2) - x**6/(6*5*4*3*2) + x**4/(4*3*2) - x**2/2 + 1
    elif F == 'sinx': 
        return x**9/(9*8*7*6*5*4*3*2) - x**7/(7*6*5*4*3*2) + x**5/(5*4*3*2) - x**3/(3*2) + x
    elif F == 'cos2x': 
        return (2*x)**8/(8*7*6*5*4*3*2) - (2*x)**6/(6*5*4*3*2) + (2*x)**4/(4*3*2) - (2*x)**2/2 + 1
    elif F == 'sin2x': 
        return (2*x)**9/(9*8*7*6*5*4*3*2) - (2*x)**7/(7*6*5*4*3*2) + (2*x)**5/(5*4*3*2) - (2*x)**3/(3*2) + 2*x
    return x**2
```

```python
x = np.arange(-3.2, 3.01, 0.01) #Для триганометрии лучше брать маленькие числа
#рисуем нашу функцию
plt.plot(x, Func(x))
plt.grid(True)


#рисуем столбцы сумм Дарбу
n = 0.3 # "Ширина столбца"
xN = np.arange(-3, 0, n) #Начало и конец промежутка суммы Дарбу



def draw_min(): #Рисуем нижнюю Дарбу
    global n, k
    k = n
    summa = 0
    xD, yD = [], [] #значения для x и y по Дарбу
    for i in xN:
        if (i >= 0): #Дарбу справа f - НЕ Триг
            if i + n > max(xN):
                k = n
                n += max(xN) - (i + n)
            xD.append(i);xD.append(i);xD.append(i+n);xD.append(i+n)
            yD.append(0);yD.append(Func(i));yD.append(Func(i));yD.append(0)
            summa += n*(math.fabs(Func(i)))
        elif (i < 0): #Дарбу слева f - НЕ Триг
            if i == min(xN):
                n = 0
            xD.append(i);xD.append(i);xD.append(i-n);xD.append(i-n)
            yD.append(0);yD.append(Func(i));yD.append(Func(i));yD.append(0)
            summa += n*(math.fabs(Func(i)))
        n = k
    print(summa)
    plt.plot(xD, yD)


def draw_max(): #Рисуем верхнюю Дарбу
    global n, k
    k = n
    summa = 0
    #print(Mn, Bn)
    xD, yD = [], [] #значения для x и y по Дарбу
    for i in xN:
        if (i >= 0): #Дарбу справа f - НЕ Триг
            if i + n > max(xN):
                k = n
                n += max(xN) - (i + n)
            xD.append(i);xD.append(i);xD.append(i+n);xD.append(i+n)
            yD.append(0);yD.append(Func(i+n));yD.append(Func(i+n));yD.append(0)
            summa += n*(math.fabs(Func(i)))
        elif (i < 0): #Дарбу слева f - НЕ Триг
            xD.append(i);xD.append(i);xD.append(i+n);xD.append(i+n)
            yD.append(0);yD.append(Func(i));yD.append(Func(i));yD.append(0)
            summa += n*(math.fabs(Func(i)))
        else:
            pass

        n = k
    print(summa)
    plt.plot(xD, yD)

#draw_min() # Нарисовать и посчитать нижнюю Дарбу
draw_max() # Нарисовать и посчитать верхнюю Дарбу
print()
plt.show()
```
